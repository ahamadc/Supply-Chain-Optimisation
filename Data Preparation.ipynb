```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data():
    order_list = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/Order List.csv")
    freight_rates = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/Freight Rates.csv")
    whcosts = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/WhCosts.csv")
    whcapacities = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/WhCapacities.csv")
    products_per_plant = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/Products per Plant.csv")
    vmi_customers = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/VMI Customers.csv")
    plant_ports = pd.read_csv("C:/Users/ahama/OneDrive/Documents/Supply Chain Data Project/Plant Ports.csv")

    tables = [order_list, freight_rates, whcosts, whcapacities, products_per_plant, vmi_customers, plant_ports]
    return tables  # Return the list of dataframes
```

```python
def inspect_df(table_list):
    for i, df in enumerate(table_list):
        print(f"Table {i+1} preview:")
        print(df.head())
        print("\nInfo:")
        print(df.info())
        print("\nStatistics:")
        print(df.describe())
        print("\n" + "="*50 + "\n")
```

```python
def clean_order_list(table_list):
    order_list = table_list[0]
    #change data type of Order Date column to date
    order_list['Order Date'] = pd.to_datetime(order_list['Order Date'], format='%d/%m/%Y')
    table_list[0] = order_list
    return(table_list)
```

```python
def clean_freight_orders(table_list):
    freight_rates = table_list[1]
    freight_rates['max_wgh_qty'] = freight_rates['max_wgh_qty'].astype(str).replace({',': '', ' ': ''}, regex = True).astype(float)
    freight_rates['minimum cost'] = freight_rates['minimum cost'].astype(str).replace({',': '', ' ': '', r'\$': ''}, regex = True).astype(float)
    freight_rates['rate'] = freight_rates['rate'].astype(str).replace({',': '', ' ': '', r'\$': ''}, regex = True).astype(float)
    table_list[1] = freight_rates
    return(table_list)
```

```python
def clean_whcosts(table_list):
    whcosts = table_list[2]
    whcosts = whcosts.dropna(subset=['WH'])
    table_list[2] = whcosts
    return(table_list)
```

```python
def clean_plant_ports(table_list):
    plant_ports = table_list[6]
    plant_ports = plant_ports[['Plant Code', 'Port']]
    table_list[6] = plant_ports
    return(table_list)
```

```python
def save_to_csv(table_list):
    # Save a DataFrame without the index column
    table_list[0].to_csv('cleaned_order_list.csv', index=False)
    table_list[1].to_csv('cleaned_freight_rates.csv', index=False)
    table_list[2].to_csv('cleaned_whcosts.csv', index=False)
    table_list[3].to_csv('cleaned_whcapacities.csv', index=False)
    table_list[4].to_csv('cleaned_products_per_plant.csv', index=False)
    table_list[5].to_csv('cleaned_vmi_customers.csv', index=False)
    table_list[6].to_csv('cleaned_plant_ports.csv', index=False)
    return(table_list)
```

```python
if __name__ == "__main__":
    tables = load_data()   # Now tables contains your list of DataFrames
    clean_order_list(tables)
    clean_freight_orders(tables)
    clean_whcosts(tables)
    clean_plant_ports(tables)
    save_to_csv(tables)
    inspect_df(tables)
```
